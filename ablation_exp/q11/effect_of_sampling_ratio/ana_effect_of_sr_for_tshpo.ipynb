```python
"""
抽样率对TSHPO的影响
"""
from tshpo.lib_class import AnaHelper

df = AnaHelper.load_csv_file("Q11_effect_of_sample_method_original_20250515_1743.csv.gz")
df
```

```python
roc_auc_df = AnaHelper.append_accuracy_of_diff_iterations_by_metric(df.query('metric=="roc_auc"'), [100], "roc_auc")
roc_auc_df
```

```python
roc_auc_plot_df = roc_auc_df.groupby(by=['metric', 'data_sample_method', 'data_sample_rate'])['metric_100'].mean()
roc_auc_plot_df = roc_auc_plot_df.reset_index()
roc_auc_plot_df
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

fig = plt.figure(figsize=(4, 2.5))
ax = fig.add_subplot(111)
sns.lineplot(roc_auc_plot_df, x="data_sample_rate", y="metric_100", ax=ax, hue='data_sample_method')
ax.set_ylim(0.94, 0.95)
# 设置ax的x轴显示0.1，0.3，0.5，0.7，1.0
ax.set_xticks([0.1, 0.3, 0.5, 0.8, 1.0])
# 设置ylabel 为 ROC AUC
ax.set_ylabel("ROC AUC")
# 设置x轴为 row sample rate
ax.set_xlabel("Row Sample Rate")
# 设置ax的legend在右下角
ax.legend(loc='lower right', title="Sampling Methods")
ax.grid(False)
# 修复图片显示不全
plt.tight_layout()
fig.savefig("Q11_effect_of_sample_method_and_sample_rate.pdf", dpi=300)
```
